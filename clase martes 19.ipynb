```python
import re

try:
    nombre_completo = input("Ingrese su nombre completo: ").lower()

    # Validar que solo haya letras y espacios
    if not re.fullmatch(r"[a-zA-Záéíóúñ\s]+", nombre_completo):
        raise ValueError("Solo se permiten letras y espacios")

    partes = nombre_completo.split()

    # Validar mínimo nombre y apellido
    if len(partes) < 2:
        raise ValueError("Debe ingresar al menos nombre y apellido")

    # Caso: un nombre y un apellido
    if len(partes) == 2:
        primer_nombre = partes[0]
        primer_apellido = partes[1]

        correo = (
            primer_nombre[0]
            + primer_apellido
            + "@gmail.com"
        )

    # Caso: dos nombres y un apellido
    elif len(partes) == 3:
        primer_nombre = partes[0]
        segundo_nombre = partes[1]
        primer_apellido = partes[1]
        segundo_apellido = partes[2]
        correo = (
            primer_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    # Caso: dos nombres y dos apellidos
    else:
        primer_nombre = partes[0]
        segundo_nombre = partes[1]
        primer_apellido = partes[2]
        segundo_apellido = partes[3]

        correo = (
            primer_nombre[0]
            + segundo_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    print("Correo generado:")
    print(correo)

except ValueError as e:
    print("Error:", e)
```

```python
import re

try:
    nombre_completo = input(
        "Ingrese sus 3 nombres y 2 apellidos: "
    ).lower()

    # Validar que solo haya letras y espacios
    if not re.fullmatch(r"[a-zA-Záéíóúñ\s]+", nombre_completo):
        raise ValueError("Solo se permiten letras y espacios")

    partes = nombre_completo.split()

    # Validar que tenga exactamente 5 palabras
    if len(partes) != 5:
        raise ValueError(
            "Debe ingresar exactamente 3 nombres y 2 apellidos"
        )

    # Asignar nombres y apellidos
    primer_nombre = partes[0]
    segundo_nombre = partes[1]
    tercer_nombre = partes[2]
    primer_apellido = partes[3]
    segundo_apellido = partes[4]

    # Crear correo
    correo = (
        primer_nombre[0]
        + segundo_nombre[0]
        + tercer_nombre[0]
        + primer_apellido
        + segundo_apellido[0]
        + "@gmail.com"
    )

    # Mostrar resultado
    print("Correo generado:")
    print(correo)

except ValueError as e:
    print("Error:", e)
```
