```python
%%writefile src/productos.py
def pedir_productos():
  productos = []
  precios = []

  for i in range(3):
    producto = input("ingrese el nombre del producto: ")
    precio = int(input("ingrese el precio del producto: "))

    productos.append(producto)
    precios.append(precio)

  return productos, precios
```

```python
%%writefile src/calculos.py
def calcular_total(precios):
  return sum(precios)

def calcular_iva(total):
  return total*0.19
```

```python
%%writefile src/factura.py
def mostrar_factura(productos, precios, total, iva):
  print("------FACTURA------")
  for i in range (len(productos)):
    print(productos[i], ":", precios[i])

  print("---------------------")
  print("total: ",total)
  print("IVA ",iva)
  print("Total a pagar: ", total + iva)
```

```python
import os
os.makedirs("empresa", exist_ok=True)
```

```python
%%writefile empresa/registro.py
def pedir_nombre():
    nombre = input("Ingrese el nombre del empleado: ")
    return nombre

def pedir_horas():
    horas = float(input("Ingrese horas trabajadas en la semana: "))
    return horas

def pedir_valor_hora():
    valor = float(input("Ingrese valor por hora: "))
    return valor
```

```python
%%writefile empresa/salario.py
def calcular_salario(horas, valor_hora):
    return horas * valor_hora

def calcular_bonificacion(salario):
    if salario > 1000000:
        return salario * 0.10
    else:
        return 0
```

```python
%%writefile empresa/reportes.py
def mostrar_empleado(nombre, horas, valor_hora):
    print("Empleado:", nombre)
    print("Horas trabajadas:", horas)
    print("Valor por hora:", valor_hora)

def mostrar_pago(salario, bono):
    print("Salario:", salario)
    print("Bonificación:", bono)
    print("Total a pagar:", salario + bono)
```

```python
from empresa import registro
from empresa import salario
from empresa import reportes

nombre = registro.pedir_nombre()
horas = registro.pedir_horas()
valor = registro.pedir_valor_hora()

sal = salario.calcular_salario(horas, valor)
bono = salario.calcular_bonificacion(sal)

reportes.mostrar_empleado(nombre, horas, valor)
reportes.mostrar_pago(sal, bono)
```
