```python
class calculadora:
  def __init__(self, a, b):
    self.a = a
    self.b = b

  def sumar(self):
    return self.a + self.b
  def multiplicar(self):
    return self.a * self.b
  def dividir(self):
    try:
      return self.a / self.b
    except ZeroDivisionError:
      return "No puede dividir por 0"
```

```python
calcula1 = calculadora(2,0)
calcula1.dividir()
```

```python
calcula2 = calculadora(float(input("Ingrese un numero a: ")), float(input("Ingrese un numero b: ")))
print("suma:",  calcula2.sumar())
print("multiplicacion:", calcula2.multiplicar())
print("division:", calcula2.dividir())
```

```python
class Estudiante:
  """Métodos: aprobar."""

  def __init__(self, nombre, nota):
    self.nombre = nombre
    self.nota = nota
  def aprobar(self):
      if self.nota >= 3:
        print(self.nombre, "Aprobó")
      else:
        print(self.nombre, "Reprobó")
```
