```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

```python
productos = {
    "pan": 1.5,
    "leche": 2.0,
    "huevos": 3.0
}
print("Productos disponibles:", productos)

try:
  nombre = input("¿Qué producto quieres?: ").lower()

  if nombre not in productos:
    raise KeyError

  cantidad = int(input("Cantidad: "))
  if cantidad <= 0:
    raise ValueError

  total = productos[nombre] * cantidad
  print("Total a pagar:", total)

except KeyError as e:
  print("Error:producto no disponible")

except ValueError as e:
  print("Error:debe ser entero")
```

```python
try:

    datos = input("Ingresa notas separadas por coma: ")

    notas = [float (n) for n in datos.split(",")]

    if len(notas) == 0:
        raise ValueError("No ingresaste notas")

    promedio = sum(notas) / len(notas)

    posicion = int(input("¿Qué nota quieres ver (índice)?: "))

    if posicion < 0:
        raise ValueError("Ingresa solo índices positivos")

    print("Nota seleccionada:", notas[posicion])

    print("Promedio:", promedio)

except ValueError as e:
    print("Error de valor:", e)

except ZeroDivisionError:
    print("No se puede dividir entre cero")

except IndexError:
    print("Índice fuera de rango")
```

permite encontrar los indices donde por primera vez aparece una cadena específica.

```python
import re
texto = "Tengo 23 perros y 3 gatos"
resultado = re.findall("\d", texto)
print(resultado)
```

re.findall("\d", texto) busca los numeros que hay en la variable y los agrupa

```python
import re
texto = "Tengo 23 perros y 3 gatos"
resultado = re.findall(r"\d+", texto)
print(resultado)
```

re.findall(r"\d+", texto) busca los digitos que hay en la variable y los agrupa

```python
import re
texto = "Tengo 23 perros y 3 gatos"
resultado = re.findall(r"\w+", texto)
print(resultado)
```

re.findall(r"\w+", texto) busca todas las palabras dentro de texto y las agrupa.
